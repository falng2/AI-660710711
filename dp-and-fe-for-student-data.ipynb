‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏Ñ‡πà‡∏≤‡∏´‡∏≤‡∏¢‡πÉ‡∏ô‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå

‡πÄ‡∏ï‡∏¥‡∏°‡∏≠‡∏≤‡∏¢‡∏∏,‡∏™‡πà‡∏ß‡∏ô‡∏™‡∏π‡∏á,‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å ‡∏ó‡∏µ‡πà‡∏´‡∏≤‡∏¢‡πÑ‡∏õ‡∏î‡πâ‡∏ß‡∏¢‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢

```python
mean_age = df_master['‡∏≠‡∏≤‡∏¢‡∏∏'].mean()
mean_height = df_master['‡∏™‡πà‡∏ß‡∏ô‡∏™‡∏π‡∏á_cm'].mean()
mean_weight = df_master['‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å_kg'].mean()
mean_hour = df_master['‡∏ä‡∏±‡πà‡∏ß‡πÇ‡∏°‡∏á‡πÄ‡∏£‡∏µ‡∏¢‡∏ô_‡∏ï‡πà‡∏≠‡∏™‡∏±‡∏õ‡∏î‡∏≤‡∏´‡πå'].mean()
df_master['‡∏≠‡∏≤‡∏¢‡∏∏'].fillna(mean_age, inplace=True)
df_master['‡∏™‡πà‡∏ß‡∏ô‡∏™‡∏π‡∏á_cm'].fillna(mean_height, inplace=True)
df_master['‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å_kg'].fillna(mean_weight, inplace=True)
df_master['‡∏ä‡∏±‡πà‡∏ß‡πÇ‡∏°‡∏á‡πÄ‡∏£‡∏µ‡∏¢‡∏ô_‡∏ï‡πà‡∏≠‡∏™‡∏±‡∏õ‡∏î‡∏≤‡∏´‡πå'].fillna(mean_hour, inplace=True)
```

‡∏ó‡∏≥‡πÉ‡∏´‡πâ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÄ‡∏õ‡πá‡∏ô‡∏°‡∏≤‡∏ï‡∏£‡∏ê‡∏≤‡∏ô‡πÄ‡∏î‡∏µ‡∏¢‡∏ß‡∏Å‡∏±‡∏ô

```python
def standardize_faculty(name):
    if name in ['‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå', '‡∏ß‡∏¥‡∏ó‡∏¢‡πå', 'Science','Sci', '‡∏Ñ‡∏ì‡∏∞‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå']:
        return '‡∏ß‡∏¥‡∏ó‡∏¢‡∏≤‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå'
    elif name in ['‡∏®‡∏¥‡∏•‡∏õ‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå', '‡∏®‡∏¥‡∏•‡∏õ‡πå', 'Arts','Liberal Arts']:
        return '‡∏®‡∏¥‡∏•‡∏õ‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå'
    elif name in ['‡∏ß‡∏¥‡∏®‡∏ß‡∏Å‡∏£‡∏£‡∏°','‡∏ß‡∏¥‡∏®‡∏ß‡∏Ø', 'Engineering','‡∏ß‡∏¥‡∏®‡∏ß‡∏∞', '‡∏ß‡∏¥‡∏®‡∏ß‡∏Å‡∏£‡∏£‡∏°‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå']:
        return '‡∏ß‡∏¥‡∏®‡∏ß‡∏Å‡∏£‡∏£‡∏°‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå'
    else:
        return name
df_master['‡∏Ñ‡∏ì‡∏∞'] = df_master['‡∏Ñ‡∏ì‡∏∞'].apply(standardize_faculty)

def standardize_sex(name):
    if name in ['‡∏ä‡∏≤‡∏¢','‡∏ä','M','Male']:
        return '‡∏ä‡∏≤‡∏¢'
    elif name in ['‡∏´‡∏ç‡∏¥‡∏á', '‡∏ç', 'F','Female']:
        return '‡∏´‡∏ç‡∏¥‡∏á'
    else:
        return name
df_master['‡πÄ‡∏û‡∏®'] = df_master['‡πÄ‡∏û‡∏®'].apply(standardize_sex)
```

‡πÄ‡∏ï‡∏¥‡∏° ‡∏Ñ‡∏ì‡∏∞,‡πÄ‡∏û‡∏® ‡∏ó‡∏µ‡πà‡∏´‡∏≤‡∏¢‡πÑ‡∏õ‡πÇ‡∏î‡∏¢‡∏≠‡∏¥‡∏á‡∏à‡∏≤‡∏Å‡∏ê‡∏≤‡∏ô‡∏ô‡∏¥‡∏¢‡∏°

```python
df_master['‡∏Ñ‡∏ì‡∏∞'].fillna(df_master['‡∏Ñ‡∏ì‡∏∞'].mode()[0], inplace=True)
df_master['‡πÄ‡∏û‡∏®'].fillna(df_master['‡πÄ‡∏û‡∏®'].mode()[0], inplace=True)
```

```python
print(df_master)
```

```python
Q1 = df_master['GPA'].quantile(0.25)
Q3 = df_master['GPA'].quantile(0.75)
IQR = Q3 - Q1
```

‡∏à‡∏±‡∏î‡∏Å‡∏≤‡∏£ GPA ‡∏ó‡∏µ‡πà‡∏ú‡∏¥‡∏î‡∏õ‡∏Å‡∏ï‡∏¥ ‡πÄ‡∏ä‡πà‡∏ô ‡∏ï‡∏¥‡∏î‡∏•‡∏ö ‡πÅ‡∏•‡∏∞‡πÄ‡∏ï‡∏¥‡∏°‡πÄ‡∏Å‡∏£‡∏î‡∏Ç‡∏≠‡∏á‡∏Ñ‡∏ô‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡∏°‡∏µ‡πÄ‡∏Å‡∏£‡∏î

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# ‡∏Å‡∏£‡∏≠‡∏á‡∏Ñ‡πà‡∏≤‡∏ú‡∏¥‡∏î‡∏õ‡∏Å‡∏ï‡∏¥
df_no_outliers = df_master[(df_master['GPA'] >= lower_bound) & (df_master['GPA'] <= upper_bound)]

print(f"\n‡∏Ç‡∏≠‡∏ö‡πÄ‡∏Ç‡∏ï‡∏•‡πà‡∏≤‡∏á - {lower_bound}")
print(f"‡∏Ç‡∏≠‡∏ö‡πÄ‡∏Ç‡∏ï‡∏ö‡∏ô - {upper_bound}")
print(f"\n‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏´‡∏•‡∏±‡∏á‡∏Å‡∏£‡∏≠‡∏á‡∏Ñ‡πà‡∏≤‡∏ú‡∏¥‡∏î‡∏õ‡∏Å‡∏ï‡∏¥")
print(df_no_outliers)
print(f"\n‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏≠‡∏≤‡∏¢‡∏∏ (‡∏´‡∏•‡∏±‡∏á‡∏Å‡∏£‡∏≠‡∏á‡∏Ñ‡πà‡∏≤‡∏ú‡∏¥‡∏î‡∏õ‡∏Å‡∏ï‡∏¥) - {df_no_outliers['GPA'].mean():.2f}")
```

‡∏•‡∏ö‡πÅ‡∏ñ‡∏ß‡∏ó‡∏µ‡πà‡∏ä‡∏∑‡πà‡∏≠‡πÄ‡∏õ‡πá‡∏ô NaN(‡πÑ‡∏°‡πà‡∏ó‡∏£‡∏≤‡∏ö)

```python
df_master_clean = df_no_outliers.dropna()
```

‡∏£‡∏ß‡∏° Features BMI,‡∏™‡∏∏‡∏Ç‡∏†‡∏≤‡∏û ‡∏£‡∏ß‡∏°‡πÄ‡∏Ç‡πâ‡∏≤‡∏Å‡∏±‡∏ö df_master_clean


```python
df_master_clean['BMI'] = df_master_clean['‡∏ô‡πâ‡∏≥‡∏´‡∏ô‡∏±‡∏Å_kg'] / (df_master_clean['‡∏™‡πà‡∏ß‡∏ô‡∏™‡∏π‡∏á_cm']/100)**2
df_master_clean['‡∏™‡∏∏‡∏Ç‡∏†‡∏≤‡∏û'] = pd.cut(df_master_clean['BMI'],
                                   bins=[0, 18.5, 25, 30, 100],
                                   labels=['‡∏ú‡∏≠‡∏°', '‡∏õ‡∏Å‡∏ï‡∏¥', '‡∏≠‡πâ‡∏ß‡∏ô', '‡∏≠‡πâ‡∏ß‡∏ô‡∏°‡∏≤‡∏Å'])

print(df_master_clean[['‡∏ä‡∏∑‡πà‡∏≠', 'BMI', '‡∏™‡∏∏‡∏Ç‡∏†‡∏≤‡∏û']])
```
